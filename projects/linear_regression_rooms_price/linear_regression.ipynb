# Linear Regression: Rooms vs Price

Simple linear regression example using synthetic data to explore the relationship between the number of rooms and housing prices.

## Description

This notebook demonstrates a basic linear regression using randomly generated data.
The goal is to understand how housing prices change as the number of rooms increases.

```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 8
```

```python
rooms = 2*np.random.randn(100,1)+3
rooms[0:50]
```

```python
price = 256 +6*rooms +abs(np.random.randn(100, 1))
price[0:50]
```

```python
plt.plot(rooms, price,  'r^')
plt.xlabel("Numbrer of rooms, 2025 Average")
plt.ylabel('2025 Average home price, 1000s USD')
plt.show()
```

```python
x = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(x,y)
print(LinReg.intercept_, LinReg.coef_)
```

```python
print(LinReg.score(x, y))
```

## Results

The linear regression model shows a positive relationship between the number of rooms and housing prices, as expected from the synthetic data generation process.


The score function returns the coefficient of determination, which is the R² of the prediction. At this point, we are getting a value of 0.997. Basically, this means that our model is performing very well, and our R² value is close to 1, which is a good thing.
